```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_validate
```

```python
df = pd.read_csv('ConditionsForNadav.csv')
df.head()
```

```python
df.columns
```

```python
count = 2000
```

```python
def contains_df(lst, new_df):
    for current_df in lst:
        if current_df.equals(new_df):
            return True
    return False
```

```python
samples = []
```

```python
for i in range(count):
    group_1, group_2 = train_test_split(df, test_size=0.5, random_state=0, stratify=df[["Body_Info", "Face_Info", "Emotion"]])
    if not contains_df(samples, group_1) and not contains_df(samples, group_2):
        samples.append(group_1)
        samples.append(group_2)
        group_1['group'] = 1
        group_2['group'] = 2
        frame = [group_1, group_2]
        group = pd.concat(frame)
        group.to_csv('samples/' + str(i) + '.csv')
    else:
        print(1)
```
